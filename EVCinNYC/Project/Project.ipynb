#### Data Collection

I don't have access to the data I need till next week (a list of users, articles they've engaged with, meta data about those articles) so I need to create some sample data.

To create this data, Google BigQuery has publicly available data on Wikipedia contributors. 
* Contributor = user
* title = title of the article
* topic = meta data about the article which I will create

I started off selecting the top 1000 titles and then looking at users that contributed to at least one of the top 1000 titles.  But there is a limit of 10,000 rows you can download in a CSV from BigQuery in the public dataset, so I fiitered my selection process to only the top 50 topics; I further had to filter my data extracts by the starting character of the user name.   The SQL I used was a join table:

SELECT u.contributor_username as username, u.title as topic, u.sessions as ses
    
    FROM
        (SELECT contributor_username, title, count(title) sessions        
        FROM [publicdata:samples.wikipedia] 
        WHERE contributor_username like 'L%'  
        GROUP EACH BY contributor_username, title) u
    JOIN
        (SELECT  title, count(distinct contributor_username) users, count(contributor_username) sessions
        FROM [publicdata:samples.wikipedia] 
        WHERE title not like 'Wikipedia%' and title not like 'Talk%' and title not like 'User%' and title not like 'Template%'
        GROUP EACH BY title
        ORDER BY users DESC
        LIMIT 50) top
ON u.title = top.title
